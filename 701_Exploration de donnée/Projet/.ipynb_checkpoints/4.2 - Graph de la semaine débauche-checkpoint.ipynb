```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from Algo_projet import *
from datetime import datetime
from datetime import time
import networkx as nx
import matplotlib.pyplot as plt
```

```python
raw_df = pd.read_csv('source/usage_stats/103JourneyDataExtract28Mar2018-03Apr2018.csv')

raw_df.drop(raw_df[raw_df['StartStation Id'] == raw_df['EndStation Id']].index,inplace=True)
raw_df.drop(raw_df[raw_df['Duration']>28800].index,inplace=True)
raw_df['Start Date'] = pd.to_datetime(raw_df['Start Date'], format="%d/%m/%Y %H:%M")
raw_df['End Date'] = pd.to_datetime(raw_df['End Date'], format="%d/%m/%Y %H:%M")
df = raw_df
df
```

```python
#Découpage du Week end
dF_Wend = df.loc[df['Start Date'].dt.strftime('%a').str.contains('^[S]')]
#une tranche horraire
dF_We_14_15 = dF_Wend.loc[(df['Start Date'].dt.time >= time(14, 0)) & (df['Start Date'].dt.time <= time(15,0))]
#dF_We_14_15['StartStation Id'].value_counts()[:10],dF_We_14_15['EndStation Id'].value_counts()[:10]
```

```python
#On détruit le graph avant
try :
    G.clear()
except :
    pass
#On construit le graph vide
G = nx.Graph(Title = "Top 10 flux 14>15")
for StartStation_Id,Freq_app_Ss in dF_We_14_15['StartStation Id'].value_counts()[:4].items():
    for EndStation_Id,Freq_app_Es in  dF_We_14_15.loc[dF_We_14_15['StartStation Id'] == StartStation_Id]['EndStation Id'].value_counts()[:10].items():
        noeud = (StartStation_Id,EndStation_Id)
        #On construit le graph par ces arrêtes
        G.add_edge(*noeud)
        #print(f"noeud : {noeud} > poids {Freq_app_Es}")
        
#affiche la liste des noeuds ou arrete           
#list(G.nodes),list(G.edges)
```

```python
#On affiche le graph (basic)
nx.draw(G, with_labels=True)
```

```python
#Fonction utilisé pour trouver le centre du graph
a = nx.k_components(G)
a
```

```python
#Affichons les points de gravité (les noeuds avec plus de 'i' connection)
a = []
grav= 1
for key,val in nx.k_components(G).items():
    if key > grav:
        a.extend(list(val[0]))
#a contient tous les noeuds les plus connexe
a = list(set(a))
#On retire tous les noeuds qui ne sont pas les plus connexe
for i in list(G.nodes):
    if i not in a :
        G.remove_node(i)
list(G.nodes)
```

```python
#Affichons le graph du centre de gravité
nx.draw(G, with_labels=True)
```
