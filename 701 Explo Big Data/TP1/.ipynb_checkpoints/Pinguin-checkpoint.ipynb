### Changement de nom : penguins
DataFrame avec donnée propre.

```python
#raw penguins sans les na
penguins=raw_penguins.dropna()
penguins
```

```python
Species = penguins["species"].unique()
Island = penguins["island"].unique()
Sex = penguins["sex"].unique()
Species,Island,Sex
```

```python
#encoder les colonnes catégorielles
from sklearn.preprocessing import LabelEncoder

Sp_le = LabelEncoder()
Is_le = LabelEncoder()
Se_le = LabelEncoder()

print (Sp_le.fit_transform(Species),Species)
print (Is_le.fit_transform(Island),Island)
print (Se_le.fit_transform(Sex),Sex)

penguins
#Sp_le.classes_, Is_le.classes_, Se_le.classes_
```

### Changement de nom : num_penguins
Var catégorielle en Var numérique

```python
num_penguins = penguins.copy()
num_penguins["species"] = Sp_le.transform(penguins["species"])
num_penguins["island"] = Is_le.transform(penguins["island"])
num_penguins["sex"] = Se_le.transform(penguins["sex"])
```

> Make a scatter plot showing bill length in function of flipper length with different colors for the different species.


```python
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(15, 10))

#[0 2 1] ['Adelie' 'Gentoo' 'Chinstrap']



for i in Species:
    x_i = num_penguins.loc[num_penguins["species"] == Sp_le.transform([i])[0]]['bill_length_mm']
    y_i = num_penguins.loc[num_penguins["species"] == Sp_le.transform([i])[0]]['flipper_length_mm']
    if i == 'Adelie' :
        c_i = 'blue'
    elif i == 'Gentoo':
        c_i = 'red'
    elif i == 'Chinstrap':
        c_i = 'green'
    else :
        c_i = 'dark'
    plt.scatter(x_i, y_i, c=c_i, alpha=0.5,label = i)


plt.title('Species')
plt.xlabel("Bill_length")
plt.ylabel("Flipper_length")
plt.legend()
plt.show()


```
